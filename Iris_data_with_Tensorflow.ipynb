```python
X=tf.placeholder('float32',[None,4])
Y=tf.placeholder('int32',[None,1])
```

```python
W=tf.Variable(tf.random_normal([4,3]))
b=tf.Variable(tf.random_normal([3]))
```

```python
Y_one_hot=tf.one_hot(Y,3)
```

```python
Y_one_hot=tf.reshape(Y_one_hot,[-1,3])
```

```python
Y_one_hot
```

```python
logits=tf.matmul(X,W)+b
hypothesis=tf.nn.softmax(logits)
```

```python
cost=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits,labels=tf.stop_gradient([Y_one_hot])))
```

```python
optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)
```

```python
prediction=tf.argmax(hypothesis,1)
correct_prediction=tf.equal(prediction,tf.argmax(Y_one_hot,1))
accuracy=tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(2001):
        c_val,ac,_=sess.run([cost,accuracy,optimizer],feed_dict={X:x_data,Y:y_data})
        if step%20==0:
            print('step:',step,
                 '\ncost:',c_val)
    
    
    print('정확성:',ac)
```
